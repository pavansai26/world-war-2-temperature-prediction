```python
z=[ElasticNet,HuberRegressor,Lasso,LinearRegression,Ridge,BaggingRegressor,ExtraTreesRegressor,GradientBoostingRegressor,
 RandomForestRegressor, AdaBoostRegressor]
```

```python
kfold=KFold(n_splits=10, shuffle=True)
```

```python
dual=[True,False]
max_iter=[100,110,120,130,140]
param_grid = dict(dual=dual)
```

```python
import time
```

```python
x=data4.iloc[1:-1].head(11000)
y=data4['MeanTemp'].head(11000)
print(y)
```

```python
for t in range(len(z)):
    grid = GridSearchCV(estimator=z[t], param_grid=param_grid, cv = kfold, n_jobs=-1)

    start_time = time.time()
    grid_result = grid.fit(x,y)
# Summarize results
    print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
    print("Execution time: " + str((time.time() - start_time)) + ' ms')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
for e in range(len(z)):
    model=z[e]
    model.fit(x_train,y_train)
    s=model.predict(x_test)
    print(mean_squared_error(y_test,s))
```
